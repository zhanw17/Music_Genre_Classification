```python
music = pd.read_csv('/content/drive/Shareddrives/ML_Final/Music_Data/Music_Genre_Data/features_30_sec.csv')
music.head()
```

```python
X = music.drop(['filename', 'length', 'label'], axis = 1)
y = music[['label']]
```

```python
print(X.columns.values)
```

```python
scaler = StandardScaler()
X_scale = scaler.fit_transform(X)
X_scale = pd.DataFrame(X_scale, columns = X.columns)
X_scale.head()
```

```python
y = music[['label']]
le = LabelEncoder()
y['label'] = le.fit_transform(y['label'])
```

```python
X_train_sc, X_test_sc, y_train, y_test = train_test_split(X_scale, y, train_size = 0.8, test_size = 0.2, random_state = 12)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, train_size = 0.8, test_size = 0.2, random_state = 12)
```

```python
def model_assess(model, X_train, X_test, y_train, y_test, title = "Default"):
    model.fit(X_train, y_train)
    preds_test = model.predict(X_test)
    preds_train = model.predict(X_train)
    #print(confusion_matrix(y_test, preds))
    print('Test Accuracy', title, ':', round(accuracy_score(y_test, preds_test), 5), '\n')
    print('Train Accuracy', title, ':', round(accuracy_score(y_train, preds_train), 5), '\n')
```

```python
# Naive Bayes
nb = GaussianNB()
model_assess(nb, X_train, X_test, y_train, y_test, "Naive Bayes")

# Stochastic Gradient Descent
sgd = SGDClassifier()
model_assess(sgd, X_train_sc, X_test_sc, y_train, y_test, "Stochastic Gradient Descent")

# KNN
knn = KNeighborsClassifier(n_neighbors=10)
model_assess(knn, X_train_sc, X_test_sc, y_train, y_test, "KNN")

# Random Forest
rforest = RandomForestClassifier()
model_assess(rforest, X_train, X_test, y_train, y_test, "Random Forest")

# Support Vector Machine
svm = SVC()
model_assess(svm, X_train_sc, X_test_sc, y_train, y_test, "Support Vector Machine")

# XGradient Booster
xgb = XGBClassifier()
model_assess(xgb, X_train, X_test, y_train, y_test, "XG Booster")
```
